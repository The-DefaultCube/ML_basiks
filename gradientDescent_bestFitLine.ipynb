```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.rcParams['figure.figsize'] = [16, 12]

import numpy as np
```

```python
x = np.array([2,3,4,5,6,7,8,9])
y = np.array([6.69, 8.6, 5.20, 11.39, 13.5, 12.3, 18.5, 14.20])

# form y = m*x + c to best fit points
# m = ?? and c = ??

# plt.plot(x,y, c = 'blue', marker='v')
plt.plot(x,y, 'o')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.axis([0,10, 0,20]) #x_min, x_max, y_min, y_max in plot
plt.show()
```

```python
def gradient_descent(x , y) :
  # find m and c so assume some initial value
  m_curr, c_curr = 0,0
  alpha = 0.02 # learning rate
  iters = 10000
  n = len(x)

  Cost_prev = 0
  for i in range(iters) :
    # find predicted values
    y_hat = m_curr*x + c_curr
    # find loss and then derivative wrt eaxh term
    # Loss = (y_hat - y)^2 --- mean sq error
    # Cost (J) = 1/n* summation of all Losses
    # now find derivative of J w.r.t m (dm) and c (dc)
    
    dm = (2/n)*sum((y_hat - y)*x)
    dc = (2/n)*sum(y_hat - y)

    m_curr -= alpha*dm
    c_curr -= alpha*dc

    ############################################
    Loss = (y_hat - y)**2
    Cost = 1/n*sum(Loss)
    # improve efficiency
    if abs(Cost_prev - Cost) <= 1e-9 :
      break

    Cost_prev = Cost
    ############################################
    print("iter : %d , m = %f, c = %f, COST = %f" %(i, m_curr, c_curr, Cost))
    
  return [m_curr, c_curr] # found m and c values

# try fixing learning rate first so cost decreses on each iteration then bump up iterations
ans = gradient_descent(x,y)  
```

```python
m,c = ans[0], ans[1]

x1 = np.linspace(0,10,100)
def f(x) :
  return m*x+c

plt.plot(x1, f(x1), color='red')

plt.plot(x,y, 'o')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.axis([0,10, 0,20])
plt.show()
```
